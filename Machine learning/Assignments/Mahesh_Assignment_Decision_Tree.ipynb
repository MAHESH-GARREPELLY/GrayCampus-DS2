```python
#importing libraries of model
from sklearn.model_selection import train_test_split
```

```python
#splitting the dataset into training set and testing set
X=data['Temperature']
Y=data['Revenue']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
X_train=np.array(X_train).reshape(-1,1)
X_test=np.array(X_test).reshape(-1,1)

Y_train=np.array(Y_train).reshape(-1,1)
Y_test=np.array(Y_test).reshape(-1,1)
```

```python
#applying decision tree regression model on training model set
from sklearn.tree import DecisionTreeRegressor
DTR=DecisionTreeRegressor()
DTR.fit(X_train,Y_train)
```

```python
#displaying training and testing scores 
print("Training score  ",DTR.score(X_train,Y_train))
print("testing score  ",DTR.score(X_test,Y_test))
```

```python
#predict the result of test set with model trained on training set and assigning to prediction variable
prediction=DTR.predict(X_test.reshape(-1,1))
prediction
```

```python
#compare and display the values of real and prediction in dataframe
df=pd.DataFrame({'Real Values':Y_test.reshape(-1),'Predicted values':prediction.reshape(-1)})
df
```

```python
#real values are plotted with blue color
#predicted values are plotted with red color
plt.scatter(X_test,Y_test,color='blue')
plt.scatter(X_test,prediction,color='red')
plt.title('Decision Tree Regression')
plt.xlabel('Temperature')
plt.ylabel('Revenue')
plt.show()
```
