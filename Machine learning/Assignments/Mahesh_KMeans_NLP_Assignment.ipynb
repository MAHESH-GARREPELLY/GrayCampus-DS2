```python
Problem Statement

The goal of text classification is to automatically classify the text documents into one or more defined categories.

Clue: You could use KNN ML algorithm to solve. Also start with concepts of NLP, then extract features from text files, then find out TF-IDF and split for classification.

End result: You have to categorize news articles into defined topics.

The dataset to be imported is fetch_20newsgroups of sklearn.
```

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#importing dataset from sklearn
from sklearn.datasets import fetch_20newsgroups
dataset=fetch_20newsgroups()
```

```python
#checking keys present in the dataset
dataset.keys()
```

```python
print(dataset['DESCR'])
```

```python
#importing tfidf vectorizer and kMeans
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
#seggregating text data and target data
dataset1=dataset.data
target=dataset.target
```

```python
#vectorizer reference for tfifdfvectorizer to fit and transfrom data
vectorizer=TfidfVectorizer(stop_words='english')
X=vectorizer.fit_transform(dataset1)
```

```python
#applying kmeans on vectorized data
clusters=10
model=KMeans(n_clusters=clusters,init='k-means++',max_iter=100,n_init=1)
model.fit(X)
```

```python
#obtaining centroids and features of clusters
order_centroids=model.cluster_centers_.argsort()[:,::-1]
terms=vectorizer.get_feature_names()
```

```python
#printing each cluster 
for i in range(clusters):
    print("clusters %d:" % i)
    for ind in order_centroids[i,:10]:
        print( '%s'  % terms[ind])
```

```python
#predicting sentensces to which cluster they belongs to 
print(" predict whether sentence belongs to cluster:")
X=vectorizer.transform(["i would like to drive car on cities"])
predicted=model.predict(X)
print(predicted)
```

```python
print(" predict whether sentence belongs to cluster:")
X=vectorizer.transform(["Amazon has free shipping"])
predicted=model.predict(X)
print(predicted)
```

```python
print(" predict whether sentence belongs to cluster:")
X=vectorizer.transform(["banks"])
predicted=model.predict(X)
print(predicted)
```
